## 🎨 Template Matching System

**NEW: Automatic Format Extraction**

The system now automatically extracts formatting from your proposals in `data/proposals/`:
- ✅ **Colors**: Extracts your RGB color palette from existing PPT files
- ✅ **Fonts**: Uses your font family, title sizes, and body text sizes
- ✅ **Layouts**: Matches slide dimensions and margins
- ✅ **Charts**: All visualizations use your color palette

**How it works:**
1. Analyzes all `.pptx` files in `data/proposals/`
2. Extracts the most common colors, fonts, and sizes
3. Applies them automatically to all exports

**To update template:**
- Add new proposal files to `data/proposals/`
- Re-run the export cells
- Template will automatically update based on new proposals
